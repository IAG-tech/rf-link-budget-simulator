## Mobile Technology Comparison — LTE800 / LTE1800 / LTE2600

This demo compares three LTE frequency bands used in mobile networks,
each paired with its most appropriate propagation model:

- **LTE800 (800 MHz) + Okumura-Hata** — valid range, good model for sub-1500 MHz urban scenarios
- **LTE1800 (1800 MHz) + COST231** — correct model for this frequency range
- **LTE2600 (2600 MHz) + COST231** — outside COST231 valid range (1500–2000 MHz), included for reference

Lower frequency bands provide greater coverage range at the cost of lower bandwidth capacity.

```python
import sys
import os
from src.run_simulation import run_simulation
sys.path.insert(0, os.path.abspath('..'))
from params.lte800_config import lte800
from params.lte1800_config import lte1800
from params.lte2600_config import lte2600
from src.utils.print_coverage import print_coverage
from dataclasses import replace
import numpy as np
import plotly.graph_objects as go
```

```python
distance = np.linspace(0.1,30,300)
lte800_Okumura = replace(lte800, name='lte800_Okumura')
lte1800_Cost231 = replace(lte1800, name='lte1800_Cost231')
lte2600_Cost231 = replace(lte2600, name='lte2600_Cost231')

[margin_lte800_Okumura, value0_lte800_Okumura,rx_lte800_Okumura] = run_simulation(lte800_Okumura,distance,'okumura-hata',False)
[margin_lte1800_cost231, value0_lte1800_cost231,rx_lte1800_cost231] = run_simulation(lte1800_Cost231,distance,'cost231',False)
[margin_lte2600_cost231, value0_lte2600_cost231,rx_lte2600_cost231] = run_simulation(lte2600_Cost231,distance,'cost231',False)
```

### Received Power vs Distance

LTE800 achieves the highest received power across all distances due to lower path loss.

LTE2600 shows the steepest decay — higher frequencies suffer greater propagation losses,

limiting coverage range significantly in urban environments.

```python
fig = go.Figure()

fig.add_trace(go.Scatter(x=distance, y=rx_lte800_Okumura, name="lte800-Okumura", line=dict(color="blue")))

fig.add_trace(go.Scatter(x=distance, y=rx_lte1800_cost231, name="lte1800-cost231", line=dict(color="red")))

fig.add_trace(go.Scatter(x=distance, y=rx_lte2600_cost231, name="lte2600-cost231", line=dict(color="green")))
fig.update_layout(
    template='plotly_dark',
    title=dict(text ='Mobile technologies Comparison', x=0.5),
    xaxis=dict(range=[0, 10]), xaxis_title='Distance (km)',
    yaxis_title='Rx power (dBm)'
)
```

### Frequency-Coverage Trade-off in Mobile Networks

| Band | Frequency | Model | Valid |
|------|-----------|-------|-------|
| LTE800 | 800 MHz | Okumura-Hata | ✓ |
| LTE1800 | 1800 MHz | COST231 | ✓ |
| LTE2600 | 2600 MHz | COST231 | ⚠️ Outside range |

In real network deployments, operators combine multiple frequency bands:
low bands for coverage and high bands for capacity in dense areas.

```python
print_coverage(lte800_Okumura.name,value0_lte800_Okumura,distance)
print_coverage(lte1800_Cost231.name,value0_lte1800_cost231,distance)
print_coverage(lte2600_Cost231.name,value0_lte2600_cost231,distance)

fig = go.Figure()

fig.add_trace(go.Scatter(x=distance, y=margin_lte800_Okumura, name="lte800-Okumura", line=dict(color="blue")))

fig.add_trace(go.Scatter(x=distance, y=margin_lte1800_cost231, name="lte1800-cost231", line=dict(color="red")))

fig.add_trace(go.Scatter(x=distance, y=margin_lte2600_cost231, name="lte2600-cost231", line=dict(color="green")))

fig.add_hline(
    y=0,
    line=dict(color='white', width=1.5, dash='dash'),
    annotation_text='Coverage threshold',
    annotation_position='top right'
)


fig.update_layout(
    template='plotly_dark',
    title=dict(text ='Mobile technologies Comparison', x=0.5),
    xaxis=dict(range=[0, 5]), xaxis_title='Distance (km)',
    yaxis_title='Link Margin (dB)',

)
```
