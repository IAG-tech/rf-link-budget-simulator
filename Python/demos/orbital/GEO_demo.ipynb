## GEO Satellite Link Budget — Dual-Band Analysis

This demo analyzes a GEO satellite link at 35,786 km altitude for two frequency bands:
- **UHF (400 MHz)** — typically used for telemetry and command (TT&C) links due to lower propagation losses
- **S-band (2 GHz)** — used for mission data downlink, offering higher bandwidth at the cost of increased path loss

Link margin and received power are evaluated as a function of elevation angle (0.1° to 90°).

```python
import sys
import os
from src.run_simulation import run_simulation
sys.path.insert(0, os.path.abspath('..'))
from params.sband_geo_config import Sband
from params.uhf_geo_config import uhf
from src.utils.print_coverage_geo import print_coverage_geo
import numpy as np
import  plotly.graph_objects as go
```

```python
angle = np.linspace(0.1,89,90)
[margin_uhf, value0_uhf,rx_uhf] = run_simulation(uhf,angle,False)
[margin_sband, value0_sband,rx_sband] = run_simulation(Sband,angle,False)
```

### Received Power vs Elevation Angle

Despite UHF having lower free-space path loss, S-band achieves higher received power
across all elevation angles due to the significantly higher gain of the parabolic dish
antenna compared to the UHF Yagi. This illustrates how antenna selection can outweigh
frequency-dependent propagation losses in satellite link design.

```python
fig = go.Figure()

fig.add_trace(go.Scatter(x=angle, y=rx_uhf, name="UHF", line=dict(color="blue")))

fig.add_trace(go.Scatter(x=angle, y=rx_sband, name="S-band", line=dict(color="red")))

fig.update_layout(
    template='plotly_dark',
    title=dict(text ='Rx power Comparison: UHF vs Sband', x=0.5),
    xaxis=dict(range=[0, 90]), xaxis_title='Angle (º)',
    yaxis_title='Rx power (dBm)'
)



```

### Link Margin vs Elevation Angle

Link margin represents the difference between received power and the minimum sensitivity threshold.
Positive values indicate viable coverage. The dashed line marks the 0 dB coverage threshold — below this, the link is not operational.

```python
print_coverage_geo(uhf.name,value0_uhf)
print_coverage_geo(Sband.name,value0_sband)

fig = go.Figure()

fig.add_trace(go.Scatter(x=angle, y=margin_uhf, name="UHF", line=dict(color="blue")))

fig.add_trace(go.Scatter(x=angle, y=margin_sband, name="S-band", line=dict(color="red")))

fig.add_hline(
    y=0,
    line=dict(color='white', width=1.5, dash='dash'),
    annotation_text='Coverage threshold',
    annotation_position='bottom right'
)


fig.update_layout(
    template='plotly_dark',
    title=dict(text ='Link Margin Comparison: UHF vs Sband', x=0.5),
    xaxis=dict(range=[0, 90]), xaxis_title='Angle (º)',
    yaxis_title='Link Margin (dB)',

)
```
