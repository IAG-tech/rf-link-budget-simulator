## Frequency Effect on Link Margin — TETRA vs LTE800

This demo isolates the effect of carrier frequency on link budget performance.

Two configurations are compared:
- **TETRA (400 MHz)** — public safety communications standard
- **LTE800 (800 MHz)** — mobile broadband at 800 MHz band

All parameters are identical except carrier frequency: same TX power, antenna gains,
cable losses and receiver sensitivity. The FSPL propagation model is used to isolate
the pure frequency effect without environmental factors.

```python
import sys
import os
from src.run_simulation import run_simulation
from src.utils.print_coverage import print_coverage
sys.path.insert(0, os.path.abspath('..'))
from params.tetra_config import tetra
from dataclasses import replace
import numpy as np
import  plotly.graph_objects as go
```

```python
distance = np.linspace(0.1,30,300)
lte800 = replace(tetra, f_MHz=800,name='LTE800')
[margin_tetra, value0_tetra,rx_tetra] = run_simulation(tetra,distance,"fspl",False)
[margin_lte, value0_lte,rx_lte] = run_simulation(lte800,distance,"fspl",False)
```

### Received Power vs Distance

Higher frequency (LTE 800 MHz) experiences greater free-space path loss at all distances,

resulting in consistently lower received power. Both curves follow the same decay shape

since only frequency differs — demonstrating the direct relationship between frequency

and propagation loss described by the Friis transmission equation.

```python
fig = go.Figure()

fig.add_trace(go.Scatter(x=distance, y=rx_tetra, name="TETRA 400MHz", line=dict(color="blue")))

fig.add_trace(go.Scatter(x=distance, y=rx_lte, name="LTE 800MHz", line=dict(color="red")))

fig.update_layout(
    template='plotly_dark',
    title=dict(text ='Rx power Comparison: TETRA vs LTE800', x=0.5),
    xaxis=dict(range=[0, 30]), xaxis_title='Distance (km)',
    yaxis=dict(range=[-90, -20]), yaxis_title='Rx power (dBm)',

)
```

### Link Margin vs Distance

Link margin quantifies the available margin above the minimum sensitivity threshold.

TETRA maintains a higher link margin across the full range due to its lower operating frequency.

The dashed line marks the 0 dB coverage threshold.

```python
print_coverage(tetra.name,value0_tetra,distance)
print_coverage(lte800.name,value0_lte,distance)
fig = go.Figure()

fig.add_trace(go.Scatter(x=distance, y=margin_tetra, name="TETRA 400MHz", line=dict(color="blue")))

fig.add_trace(go.Scatter(x=distance, y=margin_lte, name="LTE 800MHz", line=dict(color="red")))

fig.add_hline(
    y=0,
    line=dict(color='white', width=1.5, dash='dash'),
    annotation_text='Coverage threshold',
    annotation_position='bottom right'
)


fig.update_layout(
    template='plotly_dark',
    title=dict(text ='Link Margin Comparison: TETRA vs LTE800', x=0.5),
    xaxis=dict(range=[0, 30]), xaxis_title='Distance (km)',
    yaxis=dict(range=[-10, 80]), yaxis_title='Link Margin (dB)',

)
```

### Coverage Summary

Since FSPL is used without additional environmental losses, both systems show coverage

beyond 30 km — an optimistic estimate not representative of real-world deployment.

This is intentional: the purpose of this demo is to isolate the frequency effect,

not to model realistic coverage. For environment-specific analysis, see the

LTE Link Margin and LTE Technologies Comparison demos
