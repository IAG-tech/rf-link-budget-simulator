## Antenna Height Effect on Coverage — LTE1800

This demo isolates the effect of base station antenna height on link budget performance.

Three configurations are compared at 1800 MHz using the Cost231 propagation model:
- **20 m** — typical urban rooftop installation
- **40 m** — standard macro cell tower
- **80 m** — elevated rural or suburban deployment

All other parameters remain constant. Only antenna height varies.

```python
import sys
import os
from src.run_simulation import run_simulation
sys.path.insert(0, os.path.abspath('..'))
from params.lte1800_config import lte1800
from dataclasses import replace
from src.utils.print_coverage import print_coverage
import numpy as np
import  plotly.graph_objects as go
```

```python
distance = np.linspace(0.1,30,300)
lte1800_20m = replace(lte1800, hb_m=20, name='lte1800_20m')
lte1800_40m = replace(lte1800, hb_m=40, name='lte1800_40m')
lte1800_80m = replace(lte1800, hb_m=80, name='lte1800_80m')
[margin_lte1800_20m, value0_lte1800_20m,rx_lte1800_20m] = run_simulation(lte1800_20m,distance,False)
[margin_lte1800_40m, value0_lte1800_40m,rx_lte1800_40m] = run_simulation(lte1800_40m,distance,False)
[margin_lte1800_80m, value0_lte1800_80m,rx_lte1800_80m] = run_simulation(lte1800_80m,distance,False)
```

### Received Power vs Distance

Higher antenna installations achieve greater received power at all distances.

Doubling the antenna height does not produce a linear improvement —

the relationship follows the logarithmic correction factor in the Cost-Hata model.

```python
fig = go.Figure()

fig.add_trace(go.Scatter(x=distance, y=rx_lte1800_20m, name="20 m", line=dict(color="blue")))

fig.add_trace(go.Scatter(x=distance, y=rx_lte1800_40m, name="40 m", line=dict(color="red")))

fig.add_trace(go.Scatter(x=distance, y=rx_lte1800_80m, name="80 m", line=dict(color="green")))
fig.update_layout(
    template='plotly_dark',
        title=dict(text ='Antenna height Comparison', x=0.5),
    xaxis=dict(range=[0, 30]), xaxis_title='Distance (km)',
    yaxis_title='Rx power (dBm)'
)


```

### Link Margin and Coverage Distance

Coverage distance — the point where link margin crosses 0 dB — increases significantly
with antenna height:

| Height | Coverage |
|--------|----------|
| 20 m   | 1.20 km  |
| 40 m   | 1.60 km  |
| 80 m   | 2.20 km  |

Doubling the antenna height from 20 m to 40 m only produces an increase of 0.40Km.

A further doubling to 80 m increases coverage until 2.20Km — demonstrating the

non-linear but highly significant impact of antenna height in network planning.

```python
print_coverage(lte1800_20m.name,value0_lte1800_20m,distance)
print_coverage(lte1800_40m.name,value0_lte1800_40m,distance)
print_coverage(lte1800_80m.name,value0_lte1800_80m,distance)

fig = go.Figure()

fig.add_trace(go.Scatter(x=distance, y=margin_lte1800_20m, name="20m", line=dict(color="blue")))

fig.add_trace(go.Scatter(x=distance, y=margin_lte1800_40m, name="40m", line=dict(color="red")))

fig.add_trace(go.Scatter(x=distance, y=margin_lte1800_80m, name="80m", line=dict(color="green")))

fig.add_hline(
    y=0,
    line=dict(color='white', width=1.5, dash='dash'),
    annotation_text='Coverage threshold',
    annotation_position='top right'
)


fig.update_layout(
    template='plotly_dark',
    title=dict(text ='Antenna height Comparison', x=0.5),
    xaxis=dict(range=[0, 6]), xaxis_title='Distance (km)',
    yaxis_title='Link Margin (dB)',

)
```

### Key Takeaway

Antenna height is one of the most cost-effective levers in network planning.

In rural deployments where tall towers are feasible, significant coverage gains

can be achieved without increasing TX power or changing frequency —

a critical consideration for TETRA and LTE public safety network design.
