## TETRA vs LTE800 — Full System Comparison

This demo compares TETRA and LTE800 using their real technology-specific parameters:
different TX power, antenna gains, cable losses and receiver sensitivity.

Unlike the frequency-only comparison, this demo reflects a realistic deployment scenario.
The key insight: despite LTE800 operating at a higher frequency (greater path loss),
its superior receiver sensitivity compensates — resulting in nearly identical
link margin performance to TETRA across the full distance range.

```python
import sys
import os
from src.run_simulation import run_simulation
from src.utils.print_coverage import print_coverage
sys.path.insert(0, os.path.abspath('..'))
from params.tetra_config import tetra
from params.lte800_config import lte800
import numpy as np
import  plotly.graph_objects as go
```

```python
distance = np.linspace(0.1,30,300)
[margin_tetra, value0_tetra,rx_tetra] = run_simulation(tetra,distance,"fspl",False)
[margin_lte, value0_lte,rx_lte] = run_simulation(lte800,distance,"fspl",False)
```

### Received Power vs Distance

Despite TETRA operating at a lower frequency (lower path loss), LTE800 achieves

higher received power across all distances. The reason: LTE800's directional macro

antenna (15 dBi) provides 9 dB more gain than TETRA's omnidirectional antenna (6 dBi),

outweighing the propagation advantage of the lower frequency.

```python
fig = go.Figure()

fig.add_trace(go.Scatter(x=distance, y=rx_tetra, name="TETRA 400MHz", line=dict(color="blue")))

fig.add_trace(go.Scatter(x=distance, y=rx_lte, name="LTE 800MHz", line=dict(color="red")))

fig.update_layout(
    template='plotly_dark',
    title=dict(text ='Rx power Comparison: TETRA vs LTE800', x=0.5),
    xaxis=dict(range=[0, 30]), xaxis_title='Distance (km)',
    yaxis=dict(range=[-90, -20]), yaxis_title='Rx power (dBm)',

)
```

### Link Margin vs Distance

When receiver sensitivity is applied, the curves overlap significantly.
TETRA's better sensitivity (-100 dBm vs -95 dBm for LTE800) partially compensates
for its lower received power, producing comparable link margin across the full range.

This demonstrates a fundamental principle in link budget engineering:
**system performance cannot be evaluated from a single parameter** —
TX power, antenna gain, frequency and receiver sensitivity all interact
to determine the final coverage performance.


```python
print_coverage(tetra.name,value0_tetra,distance)
print_coverage(lte800.name,value0_lte,distance)
fig = go.Figure()

fig.add_trace(go.Scatter(x=distance, y=margin_tetra, name="TETRA 400MHz", line=dict(color="blue")))

fig.add_trace(go.Scatter(x=distance, y=margin_lte, name="LTE 800MHz", line=dict(color="red")))

fig.add_hline(
    y=0,
    line=dict(color='white', width=1.5, dash='dash'),
    annotation_text='Coverage threshold',
    annotation_position='bottom right'
)


fig.update_layout(
    template='plotly_dark',
    title=dict(text ='Link Margin Comparison: TETRA vs LTE800', x=0.5),
    xaxis=dict(range=[0, 30]), xaxis_title='Distance (km)',
    yaxis=dict(range=[-10, 80]), yaxis_title='Link Margin (dB)',

)
```
